**for loops**
---
A for loop allows us to iterate over an object and we can then perform and execute blocks of codes for every loop we go through. The syntax for a for loop is:

for (temporary_variable in object){
    # Execute some code at every loop
}

# For loop over a vector


```python
%%R

vec <- c(1,2,3,4,5)

for (temp_var in vec){
    print(temp_var)
}
```

```python
%%R

vec <- c(1,2,3,4,5)

for (i in 1:length(vec)){
    print(vec[i])
}
```

# For loop over a list

```python
%%R

li <- list(1,2,3,4,5)

for (temp_var in li){
    print(temp_var)
}
```

```python
%%R

li <- list(1,2,3,4,5)

for (i in 1:length(li)){
    print(li[[i]]) # Remember to use double brackets!
}
```

# For loop with a matrix

```python
%%R

mat <- matrix(1:25,nrow=5)
print(mat)
```

```python
%%R

for (num in mat){
    print(num)
}
```

# Nested for loops

```python
%%R

for (row in 1:nrow(mat)){
    for (col in 1:ncol(mat)){
        print(paste('The element at row:',row,'and col:',col,'is',mat[row,col]))
    }
}
```

```python
%%R

for (row in 1:10){
    for (col in 1:10){
        print(paste('The element at row:',row,'and col:',col,'is',row*col))
    }
}
```

# While loop

**while loops** are a while to have your program continuously run some block of code until a condition is met (made TRUE). The syntax is:

while (condition){
    # Code executed here
    # while condition is true
}

```python
# A major concern when working with a while loop is to make sure that at some point the condition should become true,
# otherwise the while loop will go forever! Remember you can use Crtl-C to kill a process in R Studio!
```

```python
%%R

print('Just a string')
```

```python
%%R

var <- 'a variable'
cat('My variable is: ',var)
```

```python
%%R

var <- 25
cat('My number is:',var)
```

```python
%%R

# Could also use:
print(paste0("Variable is: ", var))
```

```python
%%R

x <- 0

while(x < 10){

    cat('x is currently: ',x)

    # add one to x
    x <- x+1
}
```

```python
%%R

x <- 0

while(x < 10){

    cat('x is currently: ',x)

    # add one to x
    x <- x+1
    if(x==10){
        print("x is equal to 10! Terminating loop")
    }
}
```

# break

You can use break to break out of a loop. Previously we showed an if statement checking for 10, but this wasn't actually stopping the loop, the while condition check on the next run stopped the loop.

```python
%%R

x <- 0

while(x < 10){

    cat('x is currently: ',x)

    # add one to x
    x <- x+1

    if(x==10){
        print("x is equal to 10!")
        break
        print("I will also print, woohoo!")
    }
}
```

# if, else, else if Statements¶


The syntax for an if statement in R:

if (condition){
    # Execute some code
}


# else

If we want to execute another block that occurs if the if statement is false, we use an else statement to do this! It has the syntax:

if (condition) {
    # Code to execute if true
} else {
    # Code to execute if above was not true
}


```python
%%R


temp <- 75

if (temp > 80){
    print("Hot outside!")
} else if(temp<80 & temp>50){
    print('Nice outside!')
} else if(temp <50 & temp > 32){
    print("Its cooler outside!")
} else{
    print("Its really cold outside!")
}
```

## Returning Values

So far we've only been printing out results, but what if we wanted to return the results so that we could assign them to a variable.

```python
%%R

formal <- function(name='Sam',title='Sir'){
    return(paste(title,' ',name))
}
```

```python
%%R

formal()
```

```python
%%R

formal('Issac Newton')
```

```python
%%R

var <- formal('Marie Curie','Ms.')
var
```
