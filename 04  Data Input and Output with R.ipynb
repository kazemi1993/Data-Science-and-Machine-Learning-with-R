# read.table

```python
# The read.table function is the general form of read.csv,
# in fact read.csv is actually just a thin wrapper around read.table which just makes it easier to use sometimes.read.table('example.csv')
```

```python
%%R

read.table('/content/drive/MyDrive/Colab Notebooks/example.csv')
```
